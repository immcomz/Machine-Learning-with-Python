### Preprocess selected features


You should standardize your input features so the model doesn't inadvertently favor any feature due to its magnitude.
The typical way to do this is to subtract the mean and divide by the standard deviation. Scikit-learn can do this for you.


````python
from sklearn import preprocessing
column_names = ["ENGINESIZE", "FUELCONSUMPTION_COMB_MPG"]
std_scaler = preprocessing.StandardScaler()
X_std = std_scaler.fit_transform(X)

# Convert back to DataFrame for easy visualization
X_std_df = pd.DataFrame(X_std, columns=column_names)

# Print first Standardalize 4 rows in table format
print(X_std_df.head(4))

# Step 1: Calculate the Mean and Standard Deviation for Each Feature
# StandardScaler will normalize each feature (i.e., each column in the dataset)
# so that it has a mean of 0 and a standard deviation of 1.
# The formula used is:
# X_scaled = (X - μ) / σ
# Where:
# X is the value of the feature (the data point).
# μ (mu) is the mean of the feature (the average of all the values in the column).
# σ (sigma) is the standard deviation of the feature (how much the values deviate from the mean).

# This formula standardizes each feature by subtracting its mean (to center the data)
# and then dividing by the standard deviation (to scale the data).
# This results in a transformed feature with a mean of 0 and a standard deviation of 1.


# ENGINESIZE
# Mean (μ_ENGINESIZE) = (2.0 + 2.4 + 1.5 + 3.5 + 3.5 + 3.5 + 3.5 + 3.7 + 3.7) / 9 = 3.22
# Standard deviation (σ_ENGINESIZE) ≈ 0.69

# FUELCONSUMPTION_COMB_MPG
# Mean (μ_FUELCONSUMPTION) = (33 + 29 + 48 + 25 + 27 + 28 + 28 + 25 + 24) / 9 = 28.44
# Standard deviation (σ_FUELCONSUMPTION) ≈ 7.38

# CO2EMISSIONS
# Mean (μ_CO2EMISSIONS) = (196 + 221 + 136 + 255 + 244 + 230 + 232 + 255 + 267) / 9 = 227.11
# Standard deviation (σ_CO2EMISSIONS) ≈ 39.03


# Step 2: Apply the Standardization

# For ENGINESIZE:
# Formula: X_scaled_ENGINESIZE = (X_ENGINESIZE - μ_ENGINESIZE) / σ_ENGINESIZE
# Example for 2.0: (2.0 - 3.22) / 0.69 = -1.77
# Repeat for each value in ENGINESIZE.

# For FUELCONSUMPTION_COMB_MPG:
# Formula: X_scaled_FUELCONSUMPTION = (X_FUELCONSUMPTION - μ_FUELCONSUMPTION) / σ_FUELCONSUMPTION
# Example for 33: (33 - 28.44) / 7.38 = 0.62
# Repeat for each value in FUELCONSUMPTION_COMB_MPG.

# For CO2EMISSIONS:
# Formula: X_scaled_CO2EMISSIONS = (X_CO2EMISSIONS - μ_CO2EMISSIONS) / σ_CO2EMISSIONS
# Example for 196: (196 - 227.11) / 39.03 = -0.80
# Repeat for each value in CO2EMISSIONS.

# This transformation gives us a new dataset where:
# - Each feature has a mean of 0
# - Each feature has a standard deviation of 1

# Step 3: Apply the Standardization (Scaling)
# For each feature (column), we subtract the mean (μ) and divide by the standard deviation (σ).

# For ENGINESIZE feature:
# The formula is: X_scaled_ENGINESIZE = (X_ENGINESIZE - μ_ENGINESIZE) / σ_ENGINESIZE
# We apply this to each value in the ENGINESIZE column:
# Subtract the mean (3.22) and divide by the standard deviation (0.69)
# This centers and scales the data for the ENGINESIZE feature.
# Example: For ENGINESIZE = 2.0:
# X_scaled_ENGINESIZE = (2.0 - 3.22) / 0.69 = (-1.22) / 0.69 ≈ -1.77
# This means that a value of 2.0 for ENGINESIZE is about -1.77 standard deviations below the mean.

# For FUELCONSUMPTION_COMB_MPG feature:
# The formula is: X_scaled_FUELCONSUMPTION = (X_FUELCONSUMPTION - μ_FUELCONSUMPTION) / σ_FUELCONSUMPTION
# We apply this to each value in the FUELCONSUMPTION_COMB_MPG column:
# Subtract the mean (28.44) and divide by the standard deviation (7.38)
# This centers and scales the data for the FUELCONSUMPTION_COMB_MPG feature.
# Example: For FUELCONSUMPTION_COMB_MPG = 33:
# X_scaled_FUELCONSUMPTION = (33 - 28.44) / 7.38 = 4.56 / 7.38 ≈ 0.62
# This means that a value of 33 for FUELCONSUMPTION_COMB_MPG is about 0.62 standard deviations above the mean.

# For CO2EMISSIONS feature:
# The formula is: X_scaled_CO2EMISSIONS = (X_CO2EMISSIONS - μ_CO2EMISSIONS) / σ_CO2EMISSIONS
# We apply this to each value in the CO2EMISSIONS column:
# Subtract the mean (227.11) and divide by the standard deviation (39.03)
# This centers and scales the data for the CO2EMISSIONS feature.
# Example: For CO2EMISSIONS = 196:
# X_scaled_CO2EMISSIONS = (196 - 227.11) / 39.03 = (-31.11) / 39.03 ≈ -0.80
# This means that a value of 196 for CO2EMISSIONS is about -0.80 standard deviations below the mean.


````

### Create train and test datasets
Randomly split your data into train and test sets, using 80% of the dataset for training and reserving the remaining 20% for testing.


````python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X_std,y,test_size=0.2,random_state=42)
````

### Build a multiple linear regression model
Multiple and simple linear regression models can be implemented with exactly the same scikit-learn tools.


````python
from sklearn import linear_model

# create a model object
regressor = linear_model.LinearRegression()

# train the model in the training data
regressor.fit(X_train, y_train)

# Print the coefficients
coef_ =  regressor.coef_
intercept_ = regressor.intercept_

print ('Coefficients: ',coef_)
print ('Intercept: ',intercept_)

#Visualize with Normalized/Standalized(multiple independent variables 
#(Engine size and #Cylinders) ) values agains Y(CO2)
import matplotlib.pyplot as plt

# Scatter plots for each feature
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

features = ['ENGINESIZE', 'FUELCONSUMPTION_COMB_MPG']
X_train_df = pd.DataFrame(X_train, columns=features)  # Convert X_train to DataFrame

for i, feature in enumerate(features):
    axes[i].scatter(X_train_df[feature], y_train, color='blue', label="Actual data")
    
    # Predict using the feature values
    y_pred = regressor.intercept_ + regressor.coef_[0][i] * X_train_df[feature]
    axes[i].plot(X_train_df[feature], y_pred, '-r', label="Regression line")  # y = mx + c

    axes[i].set_xlabel(feature)
    axes[i].set_ylabel("CO2 Emission")
    axes[i].legend()
    axes[i].set_title(f"{feature} vs CO2 Emission")

plt.tight_layout()
plt.show()

# For visualization, we plot both features vs. y_train on the same graph.
# To display the partial effect, we "hold" the other feature at its mean (0, since the data is standardized).

# Create a figure
plt.figure(figsize=(10, 6))

# Scatter plot for ENGINESIZE vs CO2EMISSIONS
plt.scatter(X_train_df['ENGINESIZE'], y_train, color='blue', label='Data (Engine Size)')

# Calculate regression line for ENGINESIZE while holding FUELCONSUMPTION_COMB_MPG at 0
# y = intercept + coef_1 * ENGINESIZE + coef_2 * 0
x_vals_engine = np.linspace(X_train_df['ENGINESIZE'].min(), X_train_df['ENGINESIZE'].max(), 100)
y_vals_engine = intercept_ + coef_[0,0] * x_vals_engine  # partial effect for engine size
plt.plot(x_vals_engine, y_vals_engine, color='red', label='Regression Line (Engine Size)')

# Scatter plot for FUELCONSUMPTION_COMB_MPG vs CO2EMISSIONS on the same graph,
# but we shift the x-axis position so both features can be shown on a common axis.
# One approach is to offset the fuel consumption values slightly.
offset = 0.05  # small offset for visualization clarity

plt.scatter(X_train_df['FUELCONSUMPTION_COMB_MPG'] + offset, y_train, color='green', label='Data (Fuel Consumption)')

# Regression line for FUELCONSUMPTION_COMB_MPG while holding ENGINESIZE at 0:
# y = intercept + coef_2 * FUELCONSUMPTION_COMB_MPG + coef_1 * 0
x_vals_fuel = np.linspace(X_train_df['FUELCONSUMPTION_COMB_MPG'].min(), X_train_df['FUELCONSUMPTION_COMB_MPG'].max(), 100)
y_vals_fuel = intercept_ + coef_[0,1] * x_vals_fuel
plt.plot(x_vals_fuel + offset, y_vals_fuel, color='orange', label='Regression Line (Fuel Consumption)')

# Labels and title
plt.xlabel("Standardized Feature Value (with slight offset for Fuel Consumption)")
plt.ylabel("CO2 Emissions")
plt.title("Regression Lines for Standardized Features vs. CO2 Emissions")
plt.legend()
plt.show()
````

The Coefficients and Intercept parameters define the best-fit hyperplane to the data. Since there are only two variables, hence two parameters, the hyperplane is a plane. But this best-fit plane will look different in the original, unstandardized feature space. 

You can transform your model's parameters back to the original space prior to standardization as follows. This gives you a proper sense of what they mean in terms of your original input features. Without these adjustments, the model's outputs would be tied to an abstract, transformed space that doesn’t align with the actual independent variables and the real-world problem you’re solving.


````python
# Get the standard scaler's mean and standard deviation parameters
means_ = std_scaler.mean_
std_devs_ = np.sqrt(std_scaler.var_)

# The least squares parameters can be calculated relative to the original, unstandardized feature space as:
coef_original = coef_ / std_devs_
intercept_original = intercept_ - np.sum((means_ * coef_) / std_devs_)

print ('Coefficients: ', coef_original)
print ('Intercept: ', intercept_original)


````

You would expect that for the limiting case of zero ENGINESIZE and zero FUELCONSUMPTION_COMB_MPG, the resulting CO2 emissions should also be zero. This is inconsistent with the 'best fit' hyperplane, which has a non-zero intercept of 329 g/km. The answer must be that the target variable does not have a very strong linear relationship to the dependent variables, and/or the data has outliers that are biasing the result. Outliers can be handled in preprocessing, or as you will learn about later in the course, by using regularization techniques. One or more of the variables might have a nonlinear relationship to the target. Or there may still be some colinearity amongst the input variables.


### Visualize model outputs
You can visualize the goodness-of-fit of the model to the training data by plotting the fitted plane over the data. 


````python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

# Ensure X1, X2, and y_test have compatible shapes for 3D plotting
X1 = X_test[:, 0] if X_test.ndim > 1 else X_test
X2 = X_test[:, 1] if X_test.ndim > 1 else np.zeros_like(X1)

# Create a mesh grid for plotting the regression plane
x1_surf, x2_surf = np.meshgrid(np.linspace(X1.min(), X1.max(), 100), 
                               np.linspace(X2.min(), X2.max(), 100))

y_surf = intercept_ +  coef_[0,0] * x1_surf  +  coef_[0,1] * x2_surf

# Predict y values using trained regression model to compare with actual y_test for above/below plane colors
y_pred = regressor.predict(X_test.reshape(-1, 1)) if X_test.ndim == 1 else regressor.predict(X_test)
above_plane = y_test >= y_pred
below_plane = y_test < y_pred
above_plane = above_plane[:,0]
below_plane = below_plane[:,0]

# Plotting
fig = plt.figure(figsize=(20, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the data points above and below the plane in different colors
ax.scatter(X1[above_plane], X2[above_plane], y_test[above_plane],  label="Above Plane",s=70,alpha=.7,ec='k')
ax.scatter(X1[below_plane], X2[below_plane], y_test[below_plane],  label="Below Plane",s=50,alpha=.3,ec='k')

# Plot the regression plane
ax.plot_surface(x1_surf, x2_surf, y_surf, color='k', alpha=0.21,label='plane')

# Set view and labels
ax.view_init(elev=10)

ax.legend(fontsize='x-large',loc='upper center')
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
ax.set_box_aspect(None, zoom=0.75)
ax.set_xlabel('ENGINESIZE', fontsize='xx-large')
ax.set_ylabel('FUELCONSUMPTION', fontsize='xx-large')
ax.set_zlabel('CO2 Emissions', fontsize='xx-large')
ax.set_title('Multiple Linear Regression of CO2 Emissions', fontsize='xx-large')
plt.tight_layout()
plt.show()

# Predict using the trained model Not Sure the results and explanation
y_pred = regressor.predict(X_test)

# Scatter plot of actual vs predicted values
plt.scatter(y_test, y_pred, color='blue', alpha=0.5)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs. Predicted Values")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='dashed')  # Perfect fit line
plt.show()


````
